```python
from sklearn.neural_network import MLPClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
```

```python
inputs = iris.data
inputs[0:5]
```

```python
iris.feature_names
```

```python
outputs = iris.target
outputs
```

```python
iris.target_names
```

```python
inputs.shape
```

```python
iris.target.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, outputs, test_size = 0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
## Make sure to have different instances in train and test data
print(X_train[0:5])
print(X_test[0:5])
```

```python
print(y_train[0:5])
print(y_test[0:5])
```

```python
network = MLPClassifier(max_iter =1000,
                        verbose = True)
network.fit(X_train, y_train)
```

```python
network = MLPClassifier(max_iter = 2000, tol = 0.00001, verbose = True)
network.fit(X_train, y_train)
```

```python
network_logistic = MLPClassifier(max_iter = 2000, tol = 0.00001, activation = 'logistic', verbose = True, solver= 'adam',
                                learning_rate= 'constant',
                                learning_rate_init= 0.001,
                                batch_size = 32,
                                hidden_layer_sizes= (4,4))
                                 #early_stopping = True,
                                 #n_iter_no_change= 50)
network_logistic.fit(X_train, y_train)
```

```python
network_logistic.classes_
```

```python
network_logistic.intercepts_
#These are the bias weights
```

```python
network_logistic.coefs_
#These are the weights
```

```python
network_logistic.n_layers_
```

```python
network_logistic.n_outputs_
```

```python
network_logistic.out_activation_
#If we have more than 2 clases, activation function used is softmax
```

```python
predictions = network_logistic.predict(X_test)
predictions
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test, predictions)
```
